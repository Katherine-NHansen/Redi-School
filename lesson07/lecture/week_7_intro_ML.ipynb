# Introduction to Machine Learning

Machine Learning is used anywhere from automating mundane tasks to offering intelligent insights, industries in every sector try to benefit from it. 

You may already be using a device that utilizes it. For example, a wearable fitness tracker like Fitbit, or an intelligent home assistant like Google Home

__Machine learning (ML)__ is a category of an algorithm that allows software applications to become more accurate in predicting outcomes without being explicitly programmed. 

The basic premise of machine learning is to build algorithms that can receive input data and use statistical analysis to predict an output while updating outputs as new data becomes available.

### Types of Machine Learning

## Machine learning can be classified into 3 types of algorithms

1. Supervised Learning 
2. Unsupervised Learning 
3. Reinforcement Learning 

__Supervised__: 

- Here the human experts acts as the teacher where we feed the computer with training data containing the input/predictors and we show it the correct answers (output) and from the data the computer should be able to learn the patterns. 

- Computer is trained with labelled data. 


__Unsupervised__:

- The computer is trained with unlabeled data.

- Here theres no teacher at all, actually the computer might be able to teach you new things after it learns patterns in data, these algorithms a particularly useful in cases where the human expert doesnt know what to look for in the data.

- Mainly used in pattern detection and descriptive modeling. However, there are no output categories or labels here based on which the algorithm can try to model relationships. These algorithms try to use techniques on the input data to mine for rules, detect patterns, and summarize and group the data points which help in deriving meaningful insights and describe the data better to the users.


__Reinforcement__: 

- It allows machines and software agents to automatically determine the ideal behavior within a specific context, in order to maximize its performance. 

- Simple reward feedback is required for the agent to learn its behavior; this is known as the reinforcement signal

 In Supervised learning, an AI system is presented with data which is labeled, which means that each data tagged with the correct label.

The goal is to approximate the mapping function so well that when you have new input data (x) that you can predict the output variables (Y) for that data.

## Types of Supervised learning


1. Classification: A classification problem is when the output variable is a category, such as yes or no or disease and no disease.


2. Regression: A regression problem is when the output variable is a real value, such as dollars or weight.

## Dependent and Independent variables

1. Dependent variables are nothing but the variable which holds the phenomena which we are studying.\


2. Independent variables are the ones which through we are trying to explain the value or effect of the output variable (dependent variable) by creating a relationship between an independent and dependent variable.

In mathematics, this is generally explained with a formula

y=f(x)

Where,
x= independent variable
y= dependent variable

This means any changes in x will cause a change in the value of y. 

The change can be negative or positive.

This is just an association but not causality.


For example:

__We need to predict whether an employee will resign from the job or not. So, we need to analyze the various parameters that influence the mindset of employees. The possible variables or parameters which may influence are:__

 Work satisfaction

 Salary

 Distance from home and office

In the above example, we can assume that if the employee is very much satisfied with the job, getting paid with a good salary and the proximity of his/her home from the office is very low, then the probability that the employee will be resigning from his job is very low. 

The vice versa output is possible if the independent variables are just like the opposite of now. 

We are very sure that there is a relationship among these variables such that Work satisfaction, Salary, and Distance between home and office can have a possible effect and determining power on Resignation. 

Here, the variables which we use to explain or predict the outcome (Work satisfaction, Salary, Distance between home and office ) are __independent__. The one which is the main phenomenon which we are studying about  Resignation is the __dependent variable__.
